```python
import pandas as pd
import numpy as np
```

```python
houses = pd.read_csv('houses_to_rent_v2.csv')

houses.head()
```

```python
houses.describe()
```

```python
houses['bathroom']. value_counts()
```

```python
## IQR -> Distancia entre quartil (inter quantile range)
q1 = houses['bathroom']. quantile(0.25) ## Primeiro quantil
q3 = houses['bathroom']. quantile(0.75)
IQR = q3 - q1
print (f' IQR:{IQR}')
```

```python
q3 + (IQR * 1.5)
```

```python
houses_outliers = houses[(houses['bathroom']< q1 - (IQR * 1.5)) | (houses['bathroom'] > q3 + (IQR * 1.5))]
```

```python
houses_outliers
```

```python
houses_inliers = houses[(houses['bathroom']>= q1 - (IQR * 1.5)) & (houses['bathroom'] <= q3 + (IQR * 1.5))]
```

```python
houses_inliers.describe()
```
